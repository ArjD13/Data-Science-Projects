## Aim of Project: Compare yelp and foursquare api, which had the better coverage when in use for a particular region in Canada (Halifax Nova Scotia)

```python
import json
import pandas as pd
import os
import requests
from pandas.io.json import json_normalize
```

### Foursquare Api Request

```python
fs_client_id = os.environ['FOURSQUARE_CLIENT_ID']
fs_client_secret = os.environ['FOURSQUARE_CLIENT_SECRET']
```

```python
### Halifax get request for foursqaure api
hal_location = '44.6488,-63.5752'
fs_url = 'https://api.foursquare.com/v2/venues/search'
version ='20210413'
url_params = {'client_id': fs_client_id ,
              'client_secret': fs_client_secret, 
              'v': version ,
              'll': hal_location ,
              'query': 'restaurant',
              'limit': 50,
              'radius': 1000}

seafood_id = '4bb65184941ad13a093d1fe3' #id of Waterfront Warehouse Restaurant
url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(seafood_id, fs_client_id,fs_client_secret,version)
url

response = requests.get(fs_url, params=url_params)
hal_data = response.json()
```

```python
#meanigless iterations through the foursquare json response pertaining to Halifax resturant information
hal_resturants = []
hal_locations = []
hal_categories = []
hal_lat_lng = {}
for i in hal_data['response']['venues']:
    hal_resturants.append(i['name'])
    
    hal_locations.append(i['location']['formattedAddress'])
    
    for j in(i['categories']):
        hal_categories.append(j['shortName'])
    
```

### Defining information of interest and filtering a foursqaure dataframe

```python
# assign relevant part of JSON to venues
fs_data = hal_data['response']['venues']

# tranform venues into a dataframe
fs_dataframe = json_normalize(fs_data)
fs_dataframe
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in fs_dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = fs_dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python
#Making get request using foursquare api to obtain resturant ratings
ratings = {}
venue_id = []
for i in dataframe_filtered['id']:
    venue_id.append(i)



for values in venue_id:
    response = requests.get('https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(values, fs_client_id,fs_client_secret,version))
    data = response.json()
    ratings[values] = data['response']['venue']['rating'] if 'rating' in data['response']['venue'] else None
    print(ratings[values])
```

```python
ratings_signals = {}

for values in venue_id:
    response = requests.get('https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(values, fs_client_id,fs_client_secret,version))
    data = response.json()
    ratings_signals[values] = data['response']['venue']['ratingSignals'] if 'ratingSignals' in data['response']['venue'] else None
    print(ratings_signals[values])
```

```python
fs_ratingSignals_df = pd.DataFrame({'id':ratings_signals.keys(),'rating_Count': ratings_signals.values()})
```

```python
fs_ratings_df = pd.DataFrame({'id':ratings.keys(),'rating': ratings.values()})
fs_ratings_df
```

```python
#after merging ratings info this becomes the final fs dataframe for resturant information
fs_dataframe = pd.merge(pd.merge(dataframe_filtered,fs_ratings_df, on='id'),fs_ratingSignals_df, on='id')
fs_dataframe 
```

### Yelp API request

```python
yelp_ID = os.environ['yelp_ID']
yelp_key = os.environ['yelp_key']

headers = {
        'Authorization': f'Bearer {yelp_key}'
    }

# # getting restaurants in halifax using Yelp API

url_params = {
        
        'latitude': '44.6488',
        'longitude': '-63.5752',
        'term': 'restaurants',
        'limit': 50,
        'radius': 1000
    }

url = 'https://api.yelp.com/v3/businesses/search'

resp = requests.get(url=url, headers=headers, params=url_params)
data = resp.json()
```

### Yelp Dataframe

```python
# assign relevant part of JSON to venues
yelp_data = data['businesses']
# tranform venues into a dataframe
yelp_dataframe = json_normalize(yelp_data)
```

```python
yelp_dataframe.drop(columns = ['is_closed'], inplace = True)
yelp_dataframe.drop(columns = ['categories','phone','location.country'], inplace=True)
```

```python
#renaming for display purposes
renamed_columns = {}
for name in yelp_dataframe.columns:
    renamed_columns[name]= name.replace('.','_')


yelp_dataframe.rename(columns = renamed_columns)
```

### Final comparison

```python
#Yelp resturant DF w/ ratings for Halifax
yelp_dataframe.sort_values('review_count',ascending=False)[['name','review_count','rating']].head(10)
```

```python
#Foursquare resturant DF w/ ratings for Halifax
fs_dataframe.sort_values('rating_Count',ascending=False)[['name','rating_Count','rating']].head(10)
```
